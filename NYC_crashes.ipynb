```python
#The dataset contains no missing values ("" or null)
#No duplicates.
#No blank spaces (ex: " I love python " => "I love python")

class cleaning_data():
    def __init__():
        pass

    def drop_duplicates(self,df):#No duplicates.

        return df.drop_duplicates()
    
    def remove_space(self,df) :#The dataset contains no missing values ("" or null)
        return df.replace('^ | $', '', regex=True) #^.*$ means - match, from beginning to end, any character that appears zero or more times
    
    def remove_empty_rows(self,df):
        return df.dropna(how='all')
```
